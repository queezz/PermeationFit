# Backward Euler stencil

Visualize the backward Euler time-stepping stencil: the unknown at the new time level $t^{n+1}$ is coupled to the known value at $t^n$ at the same spatial point. Axis 0 = time index, axis 1 = space index.

```python
import importlib
import matplotlib.pyplot as plt
import permeation.viz.grid as grid
```

```python
# Reload grid so edits to permeation.viz.grid are picked up without restarting the kernel.
importlib.reload(grid)

# Grid: time index (rows) Ã— space index (columns). Backward Euler uses (n+1, j) and (n, j).
# mode="time" adds axis labels (time/space), semantic labels (u^{n+1}, u^n), and default title.
grid.plot_grid_stencil(
    grid_shape=(3, 5),
    center=(1, 2),
    stencil=[(0, 0), (-1, 0)],
    labels=True,
    mode="time",
)
plt.tight_layout()
plt.show()
```
